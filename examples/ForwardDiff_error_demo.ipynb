```julia
using ForwardDiff
using Plots
```

```julia
f(x) = x
g(x) = sin(π * x) # zero at every integer

function h1(x)
    h = f(x)
    h += g(x)
    return h
end

function h2(x)
    h = f(x)
    !isinteger(x) && (h += g(x))
    return h
end;
```

```julia
x = range(0,1,11)
plot(x, h1.(x))
plot!(x, h2.(x))
```

```julia
dh1(x) = ForwardDiff.derivative(h1, x)
dh2(x) = ForwardDiff.derivative(h2, x)
plot(x, dh1.(x), marker=:circle)
plot!(x, dh2.(x), marker=:circle)
```
